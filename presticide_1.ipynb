```python
#producing the graphs 
pd.options.mode.chained_assignment = None  # default='warn'

def grouping_by_product(prod_column, filter_size, *args):
    all_prods_chart1 = all_prods.copy()
    all_prods_chart1 = all_prods[all_prods.groupby(prod_column)[prod_column].transform('size')> filter_size]
        
    for filter_ in args:

        if filter_ in all_prods_chart1.columns:
            all_prods_chart1 = all_prods_chart1[all_prods_chart1[filter_]]
            break
        elif True in (all_prods['Sampling Month'].isin([filter_])):
            all_prods_chart1 = all_prods_chart1[all_prods_chart1['Sampling Month'] == filter_]
            break
        elif True in (all_prods['Retail Outlet'].isin([filter_])):
            all_prods_chart1 = all_prods_chart1[all_prods_chart1['Retail Outlet']== filter_]
            break
        elif True in (all_prods_chart1['Sampling Month'].isin([filter_])):
            all_prods_chart1 = all_prods_chart1[all_prods_chart1['Sampling Month']== filter_]
            break
        elif True in (all_prods_chart1['Country of Origin'].isin([filter_])):
            all_prods_chart1 = all_prods_chart1[all_prods_chart1['Country of Origin']== filter_]
            break
        else: 
            print('there is an error')
            break


    all_prods_chart1['Total Product Count'] = all_prods_chart1.groupby([prod_column])[prod_column].transform('count')
    all_prods_chart1['Pesticide Percentage'] = all_prods_chart1.groupby([prod_column, 'Pesticide'])['Pesticide'].transform('count')

    all_prods_chart1['Pesticide Percentage'] = all_prods_chart1['Pesticide Percentage'] /all_prods_chart1['Total Product Count']*100

    all_prods_chart1.drop_duplicates(subset = [prod_column, 'Pesticide'], inplace = True)
    for_graph=all_prods_chart1[[prod_column,'Pesticide', 'Pesticide Percentage']].copy()
    return for_graph

    
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(glove_vectors)
```

```python
import hdbscan
from hdbscan import HDBSCAN

hdbscanner = HDBSCAN()
hdbscanner.fit(X_tsne)

labels = hdbscanner.labels_
```

```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)


def scatter_2(x, y, labels, text):
    data = [
        go.Scatter(
            x=x[labels == label],
            y=y[labels == label],
            mode="markers",
            opacity=0.7,
            text=text[labels == label],
            name=label,
            marker={"size": 15, "line": {"width": 0.5, "color": "white"}},
        )
        for label in set(labels)
    ]
    layout = go.Layout(
        xaxis={"showgrid": False, "showticklabels": False, "zeroline": False},
        yaxis={"showgrid": False, "showticklabels": False, "zeroline": False},
        hovermode="closest",
    )
    fig = go.Figure(data=data, layout=layout)
    iplot(fig, config={"displayModeBar": False})


scatter_2(
    X_tsne[:, 0],
    X_tsne[:, 1],
    labels=labels.astype(str),
    text= all_prods['Description'].unique(),
)
```
